# Clustering Demonstration: Determine k using Elbow method

Here we demonstrate applying the Unsupervised Clustering method to a Randomized dataset. The choice of K is the most important parameter of the Clustering algorithm, delineating how many groups we want to separate our data. The Elbow method, where a range of K's are tested and graphed, is deployed.

For a visual demonstration of the iterative Clustering methodology, see the following links. This includes updating the points closest to the latest centroids, and updating the location of the centroids.

https://www.naftaliharris.com/blog/visualizing-k-means-clustering/
http://stanford.edu/class/ee103/visualizations/kmeans/kmeans.html


## Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

## Creating a sample dataset with 4 clusters

```python
X, y = make_blobs(n_samples=400, n_features=3, centers=4, random_state=3)
```

```python
plt.rcParams['figure.figsize'] = (16, 9)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])
```

## Use k means several times to determine k

```python
Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    Sum_of_squared_distances.append(kmeanModel.inertia_)
```

## Plot the elbow

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum of squared distances')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

## Using Elbo method, optimal cluster count is 4

```python
KMeansModel = KMeans(n_clusters=4)
KMeansModel.fit(X)
labels = KMeansModel.predict(X)
centers = KMeansModel.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels)
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], marker='*', c='#050505', s=1500)
```

## Try with cluster count 6

```python
KMeansModel = KMeans(n_clusters=6)
KMeansModel.fit(X)
labels = KMeansModel.predict(X)
centers = KMeansModel.cluster_centers_
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels)
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], marker='*', c='#050505', s=1500)
```
